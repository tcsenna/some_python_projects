## Real World Example
#### Pegar o endereço e converter para latitude e longitude (coordenadas)
#### Geocoding é pegar um engereço e passar pra coordenadas
#### Vou adicionar 2 colunas no meu dataframe: uma pra armazenar latitudes e outra para longitudes.

````python
import geopy
dir(geopy)
````

````python
from geopy.geocoders import ArcGIS
nom = ArcGIS()
````

## Geocoders pegam os endereços numa base de dados e envia pra um online service que tem todos os endereços numa base de dados e calcula a latitude e a longitude correspondente.
## Necessita de uma conexão.
## Aponto pro método geocode do objeto ArcGis e passo objeto como string
## Quando o endereço não existe, ele retorna None

````python
end1=nom.geocode('Rua Papari, 115, Bento Ribeiro, Rio de Janeiro, Brazil')
print(type(end1))
````

````python
print(end1.latitude,end1.longitude)
````

````python
nom.geocode('3995 23rd San Francisco, CA 94114')
````

````python
df=pd.read_csv('supermarkets.csv')
````

````python
df
````

````python
df.Address=df['Address']+', '+df['City']+', '+df['State']+', '+df['Country']
df
````

## Agora envio essa string pro método geocode.

````python
df['Coordinates']=df['Address'].apply(nom.geocode)
````

````python
df
````

````python
df['Coordinates']
````

````python
len(df['Coordinates'])
````

````python
df['Coordinates'][0] #mostra o texto completo pra primeira linha
````

````python
df['Coordinates'][0].latitude
````

````python
df['Latitude']=df['Coordinates'].apply(lambda x:x.latitude)
df['Longitude']=df['Coordinates'].apply(lambda x:x.longitude)
````
