- Vou transformar os dados que quero usar pra lista pq é mais fácil trabalhar com lista do que com dataframe series.

````python
lon=list(tabela.LON)
lat=list(tabela.LAT)
elev=list(tabela.ELEV)
name=list(tabela["NAME"])
````

````python
html = """
Volcano name:<br>
<a href="https://www.google.com/search?q=%%22%s%%22" target="_blank">%s</a><br>
Height: %s m
"""
mapa2 = folium.Map(location=[38.58, -99.09], zoom_start=5, tiles="Stamen Terrain")
fgv = folium.FeatureGroup(name = "Volcanoes_gotas")
def color_producer(el): #estou definindo uma cor para cada range de altitudes
    cores=['green','orange','red']
    if el<1000:
        return cores[0]
    elif el<3000:
        return cores[1]
    else:
        return cores[2]

for lt, ln, el, name in zip(lat, lon, elev, name):
    iframe = folium.IFrame(html=html % (name, name, el), width=200, height=100)
    fgv.add_child(folium.Marker(location=[lt, ln], popup=folium.Popup(iframe), icon = folium.Icon(color = color_producer(el))))
mapa2.add_child(fgv)
mapa2.save("Mapa_gotinhas.html")
mapa2
````

````python
lon=list(tabela.LON)
lat=list(tabela.LAT)
elev=list(tabela.ELEV)
name=list(tabela["NAME"])
#mapa3 = folium.Map(location=[38.58, -99.09], zoom_start=5, tiles="Stamen Terrain")
fgv2 = folium.FeatureGroup(name = "Volcanoes_circle")
for lt, ln, el, name in zip(lat, lon, elev, name):
    iframe = folium.IFrame(html=html % (name, name, el), width=200, height=100)
    fgv2.add_child(folium.CircleMarker(location=[lt, ln], radius=6, color='black', fill=True, fill_color=color_producer(el),fill_opacity=0.7,popup=folium.Popup(iframe)))
mapa2.add_child(fgv2)
mapa2.save("Mapa_bolinhas.html")
````

## Criando um mapa Choropleth:
#### Um mapa Cloropleth colore áreas que representam features .
#### Esse mapa usa esse esquema de cores para representar features.
- Quero pintar de amarelo se a população era menor que 10M, vermelho se entre 10M e 20M, e se for maior que 20M, verde.

````python
fgp = folium.FeatureGroup(name = "Population")
fgp.add_child(folium.GeoJson(data=open('world.json','r',encoding='utf-8-sig').read(),
                                  style_function= lambda x:{'fillColor':'yellow' if x['properties']['POP2005']<10000000 else 'red' if 10000000<=x['properties']['POP2005']<20000000 else 'green'}))
mapa2.add_child(fgp)
mapa2.save("Map_choropleth.html")
````

![title](choropleth.png)

### Agora vou mostrar como implementa uma feature aqui que ativa e desativa as camadas que estamos customizando.
### LayerControl procura objetos adicionados com o add_child method. Preciso de um add_child pra cada feature group, que vai ser detectado como layer na renderização.
- Para cada add_child, ele cria uma camada. Já pensou se tivesse um add_child pra cada vulcão que eu inserisse? Eu teria 62 camadas.
- Por isso é bom usar o feature_group.

````python
mapa2.add_child(folium.LayerControl())
mapa2.save("Map2_layers.html")
````

![title](layers.png)

#### You can also choose not to use feature groups actually, so you could have added GeoJSON to the map directly, however for this case for volcanoes, while you're adding, you are executing the add_child method multiple times, in that case you would have a while layer for every volcano, so you'd have added 62 children to your map object!!!!
- Se vc mapeasse um por um, vc teria 62 camadas!!!!!
- Por isso é bom agrupar em feature groups!!!
